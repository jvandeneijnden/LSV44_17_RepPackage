# Setup

```python
import matplotlib.pyplot as plt
import aplpy
from astropy import units as u

%matplotlib inline

single_col = 8.9 # cm
double_col = 18.3 # cm

def cm2inch(value):
    return value/2.54

FS = 22
LS = 22
MS = 14
MEW= 1.5

CM = 'inferno_r'
```

# Plotting both images using aplpy

Note: for this, we use a separately extracted .fits version of the casa image where the degenerate axes (i.e. the frequency and Stokes axes) have been dropped.

Note 2: the two images plotted below are combined in the paper in a single figure. In that combined figure (Combined.pdf), we have cropped the color bar of the left image (using KeyNote on Mac), as both images share the same color scale and color bar. 

## VLA Observation 8

```python
fig = plt.figure(figsize=(10,10))

data = "./FigureData/Obs8.X.dropdeg.fits"

cg = aplpy.FITSFigure(data, figure=fig, subplot=[0.15, 0.15, 0.7, 0.8])

cg.show_colorscale(cmap=CM,vmin=0.0,vmax=5e-5)

# The source coordinates
ra = 70.247207
dec = 44.530349

cg.recenter(ra, dec, radius=5./(60.*60.))  # radius is given in degrees, we want 5"
cg.show_markers(ra, dec, layer='marker_set_1', edgecolor=None, facecolor='g', marker='+', s=1000)

# Adding the beam
cg.add_beam()
cg.beam.show()
cg.beam.set_major(0.8 * u.arcsecond)
cg.beam.set_minor(0.64 * u.arcsecond)
cg.beam.set_angle(-34.43 * u.degree)
cg.beam.set_corner('bottom left')
cg.beam.set_linestyle('-')
cg.beam.set_edgecolor('k')
cg.beam.set_facecolor('w')
cg.beam.set_linewidth(2)

cg.axis_labels.set_xpad(1)
cg.axis_labels.set_ypad(0)
cg.axis_labels.set_font(size=20, weight='medium',
                         stretch='normal', family='arial',
                         style='normal', variant='normal')
cg.tick_labels.set_font(size=20, weight='medium',
                         stretch='normal', family='arial',
                         style='normal', variant='normal')
cg.ticks.set_length(6)
cg.ticks.set_linewidth(2)
cg.ticks.set_color('black')

cg.frame.set_linewidth(2)  # points
cg.frame.set_color('black')

# Define the properties of the grid
cg.add_grid()
cg.grid.show()
cg.grid.set_color('k')
cg.grid.set_linestyle('--')
cg.grid.set_linewidth(1.5)

# Colorbar:
cg.add_colorbar()
cg.colorbar.set_location('right')
cg.colorbar.set_width(0.2)
cg.colorbar.set_pad(0.3)
cg.colorbar.set_font(size=20, weight='medium', 
                      stretch='normal', family='arial', 
                      style='normal', variant='normal')
cg.colorbar.set_axis_label_text(r'Flux density ($\mu$Jy/beam)')
cg.colorbar.set_axis_label_font(size=20)
cg.colorbar.set_axis_label_pad(30)

cg.colorbar.set_label_properties()

plt.savefig('Obs8_with_colorbar.png', dpi=500)
```
