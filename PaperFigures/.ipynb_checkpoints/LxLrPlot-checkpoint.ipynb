```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as font_manager
import matplotlib.gridspec as gridspec
from matplotlib import colors
from matplotlib.patches import ConnectionPatch

from IPython.display import display, clear_output

import astropy.units as u

# If the font should not be changed, comment out the next two lines:
prop = font_manager.FontProperties(fname=path)
mpl.rcParams['font.family'] = prop.get_name()
mpl.rcParams['pdf.fonttype']=42

params = {'text.usetex':False, 'mathtext.fontset':'custom', 'mathtext.default':'regular'}
mpl.rcParams.update(params)

single_col = 8.9 # cm
double_col = 18.3 # cm
def cm2inch(value):
    return value/2.54
FS = 18
LS = 18
MS = 10
MEW= 1.5

%matplotlib inline
```

# Defining the correlation fits

```python
# The lx0 and lr0 for the Gallo et al. (2018) LMXB fits:
lx_axis = np.logspace(32,40,1000)
lx0 = 10**36.30
lr0 = 10**28.57

lr_cor_bhs = lr0 * (10**1.18) * ((lx_axis/lx0)**0.59)
lr_cor_nslmxbs = lr0 * (10**-0.17) * ((lx_axis/lx0)**0.44)
lr_cor_nshmxbs = lr0 * (10**-2.61) * ((lx_axis/lx0)**0.86)

# The limits for the LS V +44 17 fits (see LxLrFit.ipynb), using lx0 = 8.28e+36
Xi_low = 0.022131005764095598
Xi_high = 0.027605851743748386
Beta_low = 0.97
Beta_high = 1.91

lr_cor1 = lr0 * Xi_low * ((lx_axis/(8.28e+36))**Beta_low)
lr_cor2 = lr0 * Xi_high * ((lx_axis/(8.28e+36))**Beta_low)
lr_cor3 = lr0 * Xi_low * ((lx_axis/(8.28e+36))**Beta_high)
lr_cor4 = lr0 * Xi_high * ((lx_axis/(8.28e+36))**Beta_high)
    
# Plotting the outer edges of the four 1-sigma combinations
corr_min = np.minimum(np.minimum(lr_cor1, lr_cor2), np.minimum(lr_cor3, lr_cor4))
corr_max = np.maximum(np.maximum(lr_cor1, lr_cor2), np.maximum(lr_cor3, lr_cor4))
```

# Plotting

```python
### paper plot Lx Lr:

FS  = 10
LS  = 10
MS  = 6
MEW = 2

###############################################################################################################
### Setting up the figure:   
###############################################################################################################
fig = plt.figure(figsize=(cm2inch(double_col),cm2inch(10)))
ax = fig.add_subplot(111)

ax.tick_params(labelsize=LS, width=1, length=4, axis='both', which='major', pad=3)
ax.tick_params(labelsize=LS, length=2, width=1, axis='both', which='minor', pad=3)

ax.set_xscale('log')
ax.set_yscale('log')
    
ax.set_xlim(8e32, 3e39) #33
ax.set_ylim(5e25, 2.5e31) #26
ax.set_xticks([1e33, 1e34, 1e35, 1e36, 1e37, 1e38, 1e39])
ax.set_xticklabels([r'$10^{33}$', r'$10^{34}$', r'$10^{35}$', r'$10^{36}$', r'$10^{37}$', r'$10^{38}$', r'$10^{39}$'])
ax.set_yticks([1e26, 1e27, 1e28, 1e29, 1e30, 1e31])
ax.set_yticklabels([r'$10^{26}$', r'$10^{27}$', r'$10^{28}$', r'$10^{29}$', r'$10^{30}$', r'$10^{31}$'])

ax.set_xlabel('0.5-10 keV X-ray luminosity [erg/s]', fontsize=FS)
ax.set_ylabel('6 GHz radio luminosity [erg/s]', fontsize=FS)

ax.yaxis.set_ticks_position('both')
ax.xaxis.set_ticks_position('both')
ax.get_yaxis().set_tick_params(direction='in', which='both')
ax.get_xaxis().set_tick_params(direction='in', which='both')

###############################################################################################################    
### reading in and plotting the black holes:
###############################################################################################################

BH_x, BH_y = np.genfromtxt('./FigureData/BH.txt', unpack=True)
ax.plot(BH_x, BH_y, '.', c='0', ms=4, mew=1., label='Black holes',zorder=0)

###############################################################################################################
### Reading in and plotting the weak B neutron stars from Gallo et al. 2018 and Van den Eijnden et al. 2021
###############################################################################################################

NS_weakB_x, NS_weakB_y = np.genfromtxt('./FigureData/NS_weakB.txt', unpack=True)
ax.plot(NS_weakB_x, NS_weakB_y, 'o', color='0.92', ms=4, mew=1, markeredgecolor='0.4', zorder=0)

NS_weakB_2_x, NS_weakB_2_y, NS_weakB_2_dx, NS_weakB_2_dy = np.genfromtxt('./FigureData/NS_weakB_2.txt', unpack=True)
ax.errorbar(NS_weakB_2_x, NS_weakB_2_y, xerr=NS_weakB_2_dx, yerr=NS_weakB_2_dy,
            fmt='o', color='0.92', ms=4, mew=1., markeredgecolor='0.4', label=r'NS LMXBs',zorder=0)

###############################################################################################################    
### Plotting the strong B NSs in persistent HMXBs
###############################################################################################################
     
NS_persHMXB_x, NS_persHMXB_y, NS_persHMXB_dx, NS_persHMXB_dy = np.genfromtxt('./FigureData/NS_persistent_HMXBs.txt', unpack=True)
ax.errorbar(NS_persHMXB_x, NS_persHMXB_y, xerr=NS_persHMXB_dx, yerr=NS_persHMXB_dy,
            fmt='s', color='0.7', ms=4, mew=1., markeredgecolor='k', ecolor='k', elinewidth=1, zorder=1,
            label='Persistent NS HMXBs')

###############################################################################################################    
### Plotting the transient BeXRBs
###############################################################################################################

SwJ0243_UL_x, SwJ0243_UL_y, SwJ0243_UL_dx, SwJ0243_UL_dy = np.genfromtxt('./FigureData/SwJ0243_ULs.txt', unpack=True) 
ax.errorbar(SwJ0243_UL_x, SwJ0243_UL_y, xerr=SwJ0243_UL_dx, yerr=SwJ0243_UL_dy, 
            fmt='o', color='m', ms=0, mew=1., mec='k', uplims=True,zorder=5.,ecolor='k')
ax.plot(SwJ0243_UL_x, SwJ0243_UL_y, 'sm', ms=5, mew=1., mec='k', zorder=5.)

    
SwJ0243_x, SwJ0243_y, SwJ0243_dx, SwJ0243_dy = np.genfromtxt('./FigureData/SwJ0243.txt', unpack=True) 
ax.errorbar(SwJ0243_x, SwJ0243_y, xerr=SwJ0243_dx, yerr=SwJ0243_dy, 
            fmt='s', color='m', ms=5, mew=1., markeredgecolor='k',
            label='Swift J0243.6+6124')

A0535_x, A0535_y, A0535_dx, A0535_dy, A0535_lim = np.genfromtxt('./FigureData/A0535.txt', unpack=True) 
ax.errorbar(A0535_x, A0535_y, xerr=A0535_dx, yerr=A0535_dy, 
            fmt='b8', mew=1., mec='k', ms=5, elinewidth=1., uplims=A0535_lim, ecolor='k', zorder=3,
            label='1A 0535+262')

LS_V_44_17_x, LS_V_44_17_y, LS_V_44_17_dy, LS_V_44_17_lim = np.genfromtxt('./FigureData/LS_V_44_17.txt', unpack=True) 
ax.errorbar(LS_V_44_17_x, LS_V_44_17_y, yerr=LS_V_44_17_dy, 
            fmt='r8', mew=1.5, mec='k', ms=6, elinewidth=1.5, uplims=LS_V_44_17_lim, zorder=10, ecolor='k',\
            label=r'LS V +44 17')

###############################################################################################################
### Plotting correlations
###############################################################################################################

ax.plot(lx_axis, (6./5.)*lr_cor_bhs, 'k--', lw=0.75)
ax.plot(lx_axis, (6./5.)*lr_cor_nslmxbs, '--', color='0.4', lw=0.75)
ax.plot(lx_axis, lr_cor_nshmxbs, '--', color='b', lw=0.75)
        
# LS V +44 17    
lr_cor_plot = lr0 * (0.025644908348372547) * ((lx_axis/(8.0580490725e+36))**1.25)
ax.plot(lx_axis, lr_cor_plot, '--', color='r', lw=1.5)
ax.plot(lx_axis, corr_min, '-', color='r', lw=0.5)
ax.plot(lx_axis, corr_max, '-', color='r', lw=0.5)
ax.fill_between(lx_axis, corr_min, corr_max, color='r', alpha=0.1)

# the other to BeXRBs: based on Van den Eijnden et al. (2022)
log_model_axis = np.linspace(30,40,1000,endpoint=True)
model_axis = 10.0**log_model_axis

case1 = 4.198556386987543e+27 * ((model_axis/(2.0166088838527146e+38))**(0.855-0.057)) * (10**(0.057+0.052))
case2 = 4.198556386987543e+27 * ((model_axis/(2.0166088838527146e+38))**(0.855+0.057)) * (10**(0.057-0.052))
case3 = 4.198556386987543e+27 * ((model_axis/(2.0166088838527146e+38))**(0.855-0.057)) * (10**(0.057-0.052))
case4 = 4.198556386987543e+27 * ((model_axis/(2.0166088838527146e+38))**(0.855+0.057)) * (10**(0.057+0.052))
maxline = []
minline = []
for i in range(len(case1)):
    maxline.append(max(case1[i], case2[i], case3[i], case4[i]))
    minline.append(min(case1[i], case2[i], case3[i], case4[i]))

ax.fill_between(model_axis, minline, maxline, facecolor='b', alpha=0.1)
ax.plot(model_axis, minline, 'b', lw=0.5)
ax.plot(model_axis, maxline, 'b', lw=0.5)

###############################################################################################################
### Making the legend:
###############################################################################################################

### The legend:
leg = ax.legend(loc=2, frameon=True, fontsize=8.5, ncol=2, numpoints=1, fancybox=False, framealpha=1.0)
leg.get_frame().set_edgecolor('k')
leg.get_frame().set_linewidth(0.7)

###############################################################################################################
plt.tight_layout()
plt.savefig('./LxLr_new_noJ1008.pdf', dpi=300)
```
